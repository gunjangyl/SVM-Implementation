```python
#Import Essential libraries
import numpy as np
import pandas as pd
```

```python
#Read Dataset
df=pd.read_csv("dtsvm.csv")
df.head()
```

```python
#Separate dataset into independent & dependent variables
#Extracting independent variable X i.e. Gender & Age
X=df.iloc[:,[2,3]].values
X
```

```python
#Extracting dependent variable y i.e Purchased
y=df.iloc[:,4].values
y
```

```python
#Checking for missing values
df.isnull().sum() #checking if missing values exist or not
```

```python
#Handling missing values
from sklearn.impute import SimpleImputer
si=SimpleImputer(missing_values=np.nan, strategy='mean')
X=si.fit_transform(X)
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X_test
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X_train=ss.fit_transform(X_train)
X_test=ss.fit_transform(X_test)
```

```python
#Applying SVM(Support Vector Machine) Algorithm 
from sklearn.svm import SVC #SVC-SupportVectorClassifier
svc=SVC(kernel="linear", random_state=0)
svc.fit(X_train, y_train)
```

```python
#Do testing of model with testing dataset
y_pred=svc.predict(X_test)
```

```python
y_pred #predicted values
```

```python
y_test #original(test) values
```

```python
#Matching predicted & test values
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test, y_pred)
cm 
```
